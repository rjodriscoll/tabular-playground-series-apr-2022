```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("../Data/train_features_computed_tabular.csv")
labels = pd.read_csv("../Data/train_labels.csv")
data = data.merge(labels, on="sequence", how="left")
X, y = data.drop(["sequence", "state"], axis=1).values, data["state"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from tpot import TPOTClassifier
```

```python
pipeline_optimizer = TPOTClassifier(generations=100, population_size=100, cv=5, scoring = 'roc_auc',
                                    random_state=42, verbosity=2, max_time_mins = 120)
```

```python
pipeline_optimizer.fit(X_train, y_train)
```

```python
print(pipeline_optimizer.score(X_test, y_test))
```
